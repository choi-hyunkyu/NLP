### 텍스트 파일 읽어오기

```python
import re
import pandas as pd


with open('./data/KCC940_Korean_sentences_UTF8.txt', 'r', encoding = 'utf-8') as text_file:
    sentence_list = text_file.readlines()
    
fixed_sentence_list = []
for sentence in sentence_list:
    index = re.compile('[^ ㄱ-ㅣ가-힣]+') # 한글과 띄어쓰기를 제외한 모든 글자
    parse = index.sub(' ', sentence)
    fixed_sentence_list.append(parse)
```
