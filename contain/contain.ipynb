### 텍스트 파일 읽어오기

```python
import re

with open('./data/KCC940_Korean_sentences_UTF8.txt', 'r', encoding = 'utf-8') as text_file:
    sentence_list = text_file.readlines()
    
fixed_sentence_list = []
for sentence in sentence_list:
    index = re.compile('[^ ㄱ-ㅣ가-힣]+') # 한글과 띄어쓰기를 제외한 모든 글자
    parse = index.sub(' ', sentence)
    fixed_sentence_list.append(parse)
```

### 출력

```python
value = "안녕"
```

#### 단어 포함

```python
matching_word = [string for string in fixed_sentence_list if value in string]
```

#### 문장 앞 부분 단어 포함

```python
matching_front = [string for string in fixed_sentence_list if value in string[:len(value)]]
```

#### 문장 뒷 부분 단어 포함

```python
matching_back = [string for string in fixed_sentence_list if value in string[len(string)-len(value) - 1:]]
```
